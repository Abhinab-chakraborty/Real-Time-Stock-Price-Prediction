```python
import json
import time
import yfinance as yf
from kafka import KafkaProducer

# Kafka configuration
KAFKA_TOPIC = 'stck-data'
KAFKA_BROKER = 'localhost:9092'  # Change if your broker is different

# Create a Kafka producer
producer = KafkaProducer(
    bootstrap_servers=KAFKA_BROKER,
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

def fetch_and_send_stock_data(symbol):
    stock = yf.Ticker(symbol)
    data = stock.history(period='1d')  # Fetch the latest day's data
    for index, row in data.iterrows():
        message = {
            'symbol': symbol,
            'timestamp': index.isoformat(),
            'price': row['Close']
        }
        producer.send(KAFKA_TOPIC, value=message)
        print(f"Sent: {message}")
        time.sleep(1)  # Sleep to avoid hitting the API too quickly

if __name__ == "__main__":
    # List of stock symbols to fetch data for
    stock_symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA', 'NFLX']  # Add more symbols as needed

    while True:
        for stock_symbol in stock_symbols:
            fetch_and_send_stock_data(stock_symbol)
        time.sleep(60)  # Sleep for a minute before fetching data again for all stocks
```
